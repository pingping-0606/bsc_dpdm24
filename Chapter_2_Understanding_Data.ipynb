```python
# prompt: read file json 2298

import json
import pandas as pd

# Assuming 'json_file_path' is defined in your environment
json_file_path = '/content/drive/MyDrive/bsc_dpdm_me/memo_2298.json'

try:
    with open(json_file_path, 'r') as f:
        json_data = json.load(f)

    # Check if the JSON data is a list of dictionaries
    if isinstance(json_data, list) and all(isinstance(item, dict) for item in json_data):
      df_json = pd.DataFrame(json_data)
    elif isinstance(json_data, dict):
      df_json = pd.DataFrame([json_data])  # Create a DataFrame from a single dictionary
    else:
      print(f"Error: Invalid JSON data structure. Expected a list of dictionaries or a dictionary.")
      df_json = pd.DataFrame() # Empty DataFrame as default

    print(df_json.head())

except FileNotFoundError:
    print(f"Error: File not found at {json_file_path}")
except json.JSONDecodeError:
    print(f"Error: Invalid JSON format in the file at {json_file_path}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
