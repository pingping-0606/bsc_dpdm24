```python
# Count the number of unique continents
num_continents = Data['Continent'].nunique()
print(f"Number of unique continents: {num_continents}")

# Get a list of unique continents
unique_continents = Data['Continent'].unique()
print("Unique Continents:")
for continent in unique_continents:
    print(continent) # Indent this line to print each continent within the loop
```

```python
# Remove rows where the continent is 'Unknown'
Data_new = Data[Data['Continent'] != 'Unknown']
Data_new
```

```python
# prompt: เช็คว่าข้อมูลเหลือกี่ %

# Calculate the percentage of remaining data
remaining_percentage = (len(Data_new) / len(Data)) * 100
print(f"Percentage of data remaining after removing 'Unknown' continents: {remaining_percentage:.2f}%")
```

```python
# prompt: เเต่ละทวีปมีลุกค้ากี่คน จำนวนครั้งที่ซื้อ  จำนวนสินค้าเฉลี่ยต่อใบเสร็จ(เลขจำนวนเต็ม) ยอดขายเฉลี่ยต่อใบเสร็จ(2ตำเเหน่ง)

# Group data by continent and customer ID
continent_customer_analysis = Data_new.groupby('Continent')['CustomerID'].agg(['nunique', 'count'])
continent_customer_analysis = continent_customer_analysis.rename(columns={'nunique': 'Number of Customers', 'count': 'Number of Purchases'})


# Calculate average quantity and average sales per invoice
average_metrics = Data_new.groupby('Continent').agg(
    Average_Quantity=('Quantity', 'mean'),
    Average_Sales=('UnitPrice', 'mean')
)

# Round average quantity to the nearest integer
average_metrics['Average_Quantity'] = average_metrics['Average_Quantity'].round().astype(int)

# Format average sales to two decimal places
average_metrics['Average_Sales'] = average_metrics['Average_Sales'].round(2)

# Combine the results
result = pd.merge(continent_customer_analysis, average_metrics, on='Continent')

# Display the results
result
```

```python
# prompt: แต่ละทวีปซื้อสินค้าเยอะสุด 10 อันดับแรก   ใช้Association Rules

# Assuming 'Data_new' DataFrame is already created and contains 'InvoiceNo', 'Description', and 'Continent' columns.

# Group transactions by continent and item description
continent_item_counts = Data_new.groupby(['Continent', 'Description'])['CustomerID'].count().reset_index(name='CustomerCount')

# Sort by continent and transaction count in descending order to get the top 10 items for each continent
top_items_by_continent = continent_item_counts.groupby('Continent').apply(lambda x: x.nlargest(10, 'CustomerCount')).reset_index(drop=True)
top_items_by_continent
```

## 2.ในทั้ง 7 (Monday Tuesday Wednesday Thursday Friday Saturday Sunday)ของทวีปของชนิดใดขายคู่กันดีที่สุดและอยู่ในวันใด

```python
# prompt: แบ่ง InvoiceDate เป็นวันทั้ง 7 และเพิ่มเข้าไปในตารางที่ CustomerID แล้ว

# Assuming that 'Europe_data' is already defined and loaded.
import pandas as pd

# Convert 'InvoiceDate' to datetime objects
Europe_data['InvoiceDate'] = pd.to_datetime(Europe_data['InvoiceDate'])

# Extract the day of the week (0=Monday, 6=Sunday)
Europe_data['DayOfWeek'] = Europe_data['InvoiceDate'].dt.dayofweek

# Create a dictionary to map dayofweek to day names
day_names = {
    0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday',
    4: 'Friday', 5: 'Saturday', 6: 'Sunday'
}

# Map dayofweek to day names in the dataframe
Europe_data['DayName'] = Europe_data['DayOfWeek'].map(day_names)

# Group by CustomerID and DayName, then aggregate the items
customer_day_items = Europe_data.groupby(['CustomerID', 'DayName'])['StockCode'].apply(list).reset_index()

# Now you have a DataFrame where each row represents a customer, a day of the week, and the list of items they purchased on that day
customer_day_items
```

```python
# prompt: จากcustomer_day_items วันเสาร์มีไหม

# Assuming 'customer_day_items' has been defined previously

saturday_exists = 'Saturday' in customer_day_items['DayName'].unique()

if saturday_exists:
    print("Saturday exists in the dataset.")
    saturday_data = customer_day_items[customer_day_items['DayName'] == 'Saturday']
    print(saturday_data)  # Print data for Saturday
else:
    print("Saturday does not exist in the dataset.")
```

```python
# prompt: ตาราง customer_day_items จากนั้น groupby DayName

# Assuming 'customer_day_items' has been defined previously

day_items = customer_day_items.groupby('DayName')['StockCode'].apply(list)
day_items  # Display the grouped data
```

```python
# prompt: ดูแยกว่าแต่ละวัน StockCode อันไหนคู่กันมากสุด

# Assuming 'day_items' is already defined as in the provided code.

# Apply apriori algorithm for each day
from apyori import apriori

for day, items in day_items.items():
    # Flatten the list of lists into a single list of items
    flattened_transactions = [item for sublist in items for item in sublist]

    # Convert the flattened list into a list of transactions, each a list of items
    transactions = [list(set(sublist)) for sublist in items]

    # Apply apriori
    if transactions :
      frequent_itemsets = list(apriori(transactions, min_support=0.07))  # Adjust min_support as needed

      print(f"\nAssociation Rules for {day}:")
      for itemset in frequent_itemsets:
          if len(itemset[0]) > 1 :  # Print rules with more than one item
              print(itemset) # Indented this line to print the itemset when the condition is True
    else:
        print(f"\nNo transactions found for {day}")
```
